```python
import matplotlib.pyplot as plt
```

```python
x1 = [1, 2, 3, 4]
y1 = [45, 10, 25, 88]
a = [1, 2, 3, 4]
b = [2, 5, 25, 49]
plt.plot(x1, y1, marker="o")
plt.plot(a, b, marker="o")
plt.title("Line Plot")
plt.legend(["line1", "line2"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.plot(x1)
```

```python
plt.plot(x1,y1,color="red",linestyle="--",marker="o",markerfacecolor="green",linewidth="2")
```

```python
plt.bar(a, b)
plt.title("Bar Plot")
plt.legend(["line1"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.scatter(a, b)
plt.title("Scatter Plot")
plt.legend(["line1"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
l=[0,0,0.1,0]
plt.pie(a,explode=l)
plt.title("Pie Plot")
plt.legend(["a","b","c","d"])
```
